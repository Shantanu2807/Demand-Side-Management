```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
# Load Plant 1 Generation Data
plant1_generation_data = pd.read_csv('C:/Users/shant/OneDrive/Desktop/minor/DATASETS/Plant_1_Generation_Data.csv')

# Load Plant 1 Weather Sensor Data
plant1_weather_sensor_data = pd.read_csv('C:/Users/shant/OneDrive/Desktop/minor/DATASETS/Plant_1_Weather_Sensor_Data.csv')
```

```python
# Handle missing values
plant1_generation_data.dropna(inplace=True)
plant1_weather_sensor_data.dropna(inplace=True)

# Drop duplicates
plant1_generation_data.drop_duplicates(inplace=True)
plant1_weather_sensor_data.drop_duplicates(inplace=True)
```

```python
plant1_generation_data.info()
```

```python
plant1_weather_sensor_data=plant1_weather_sensor_data.iloc[:34]
plant1_weather_sensor_data.info()
```

```python
plant1_generation_data['TOTAL_YIELD']=plant1_generation_data['TOTAL_YIELD']/1000
plant1_generation_data.info()
```

```python
# Convert DATE_TIME column to datetime data type
plant1_generation_data['DATE_TIME'] = pd.to_datetime(plant1_generation_data['DATE_TIME'])

# Group by date and calculate the mean daily yield
average_daily_yield_plant1 = plant1_generation_data.groupby(plant1_generation_data['DATE_TIME'].dt.date)['DAILY_YIELD'].mean()
daily_yields = average_daily_yield_plant1.values

# Display  average daily yield
print(average_daily_yield_plant1)
```

```python
# Calculate DC Power statistics for Plant 1
plant1_dc_power_stats = plant1_generation_data['DC_POWER'].describe()

print("Plant 1 DC Power Statistics:")
print(plant1_dc_power_stats)
```

```python
# Group data by date and calculate total AC power generated for each day
ac_power_per_day_plant1 = plant1_generation_data.groupby(plant1_generation_data['DATE_TIME'].dt.date)['AC_POWER'].sum()

# Create a bar graph
plt.figure(figsize=(12, 6))
ac_power_per_day_plant1.plot(kind='bar', color='orange')
plt.title('Total AC Power Generated per Day for Plant 1')
plt.xlabel('Date')
plt.ylabel('Total AC Power Generated (kW)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(plant1_weather_sensor_data['AMBIENT_TEMPERATURE'], plant1_weather_sensor_data['MODULE_TEMPERATURE'],alpha=0.5)
plt.title("Correlation between Plant 1's Ambient Temperature and Module Temperature")
plt.xlabel("Ambient Temperature (°C)")
plt.ylabel("Module Temperature (°C)")
plt.grid(True)
plt.show()
```

```python
from sklearn.neighbors  import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,mean_squared_error
```

```python
df = pd.DataFrame({'MODULE_TEMPERATURE': plant1_weather_sensor_data['MODULE_TEMPERATURE'],
                   'DAILY_YIELD': daily_yields,
                   'TIME':plant1_weather_sensor_data['DATE_TIME']})
x=df['MODULE_TEMPERATURE'].to_numpy()
y=df['DAILY_YIELD'].to_numpy()
x=x.reshape(-1, 1)
y=y.reshape(-1, 1)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
df.head()
```

```python
rfr=KNeighborsRegressor(n_neighbors=5)
rfr.fit(x_train,y_train)
y_pred=rfr.predict(x_test)
accuracy=mean_squared_error(y_test,y_pred)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x_test, y_test, color='blue', alpha=0.5, label='Actual Data')
plt.plot(x_test, y_pred, color='red', linewidth=2, label='Regression Line')
plt.title('KNN Regression: Module Temperature vs. Total Yield')
plt.xlabel('Module Temperature (°C)')
plt.ylabel('Total Yield')
plt.legend()
plt.grid(True)
```
